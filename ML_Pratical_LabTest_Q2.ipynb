```python
Step 1: Import Libraries
import os
import cv2
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_path = "chest_xray/train"

normal_count = len(os.listdir(train_path + "/NORMAL"))

pneumonia_count = len(os.listdir(train_path + "/PNEUMONIA"))

print("Normal Images:", normal_count)

print("Pneumonia Images:", pneumonia_count)
```

```python
# Step 3: Dataset Profiling
print("Normal:", normal_count)

print("Pneumonia:", pneumonia_count)

print("Total Images:",
      normal_count + pneumonia_count)
```

```python
for category in ["NORMAL","PNEUMONIA"]:

    path = train_path + "/" + category

    for img in os.listdir(path):

        try:

            cv2.imread(
                os.path.join(path,img)
            )

        except:

            print(img)
```

```python
plt.figure(figsize=(10,5))

for i,img_name in enumerate(
    os.listdir(train_path+"/NORMAL")[:5]
):

    img = cv2.imread(
        os.path.join(
            train_path+"/NORMAL",
            img_name
        )
    )

    plt.subplot(1,5,i+1)

    plt.imshow(img)

    plt.axis("off")

plt.show()
```

```python
# Show Normal Image
img = cv2.imread(
    train_path+"/NORMAL/" +
    os.listdir(
        train_path+"/NORMAL"
    )[0]
)

plt.imshow(img,cmap='gray')

plt.title("Normal")

plt.axis("off")

plt.show()
```

```python
# Show Pneumonia Image
img = cv2.imread(
    train_path+"/PNEUMONIA/" +
    os.listdir(
        train_path+"/PNEUMONIA"
    )[0]
)

plt.imshow(img,cmap='gray')

plt.title("Pneumonia")

plt.axis("off")

plt.show()
```

```python
# Class Distribution Plot
sns.barplot(

    x=['Normal','Pneumonia'],

    y=[normal_count,pneumonia_count]
)

plt.show()
```
